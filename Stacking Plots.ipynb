```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
path = os.path.join(os.getcwd(),"MagTests")
os.listdir(path)
files = list(filter(lambda eachFile: os.path.splitext(eachFile)[1] == ".csv", os.listdir(path)))
files
```

```python
filepaths = []
for eachFile in files:
    filepaths.append(os.path.join(path,eachFile))
filepaths
```

```python
i = 0
lists_of_data = []
for eachFile in filepaths:
    try:
        df = pd.read_csv(eachFile, header=0)
        lists_of_data.append(df['Potential (V)'].to_list())
    except Exception as e:
        print(e)

[print(x) for x in lists_of_data]
    
    
```

```python
x_values = np.linspace(0,10,1000)
```

```python
j = 0
for eachDataSet in lists_of_data:
    plt.plot(x_values, eachDataSet, label = f"Test {j}")
    j += 1

plt.xlabel("time (s)")
plt.ylabel("potential (V)")
plt.legend()
plt.title("comparison of mag tests over 10 second intervals sampled every 10ms")
plt.savefig("Comparison of Mag Tests.png")
plt.show()
```
