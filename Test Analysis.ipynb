```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec 
%matplotlib inline
```

```python
path = os.path.join(os.getcwd(),"Tests")
os.listdir(path)
```

```python
df = pd.read_csv(os.path.join(path,"2020-06-29_15-18_large_disc.csv"),header=0)
isSignificant = abs(df['Potential (V)']) > 0.05
df_filtered = df[isSignificant]
df_filtered
```

```python

f = plt.figure(figsize=(20,10))
gs = gridspec.GridSpec(1,2,width_ratios=[10,1])

# First plot
ax1 = plt.subplot(gs[0])
ax1.set_title("Large Piezo Test (sampled every 10ms)")
ax1.set_xlabel("Seconds")
ax1 = sns.lineplot(x=np.linspace(0,10,1000), y = df['Potential (V)'])

# Second plot
ax2 = plt.subplot(gs[1])
ax2.set_title("Violin plot of sigificant values (>|0.05V|)")
ax2.set_ylabel("Potential (V)")
ax2 = sns.violinplot(y=df_filtered['Potential (V)'])

plt.show()
```

```python
df2 = pd.read_csv(os.path.join(path,"2020-06-29_14-38_small_disc_test.csv"),header=0)
isSignificant = abs(df2['Potential (V)']) > 0.05
df2_filtered = df2[isSignificant]
df2_filtered

f = plt.figure(figsize=(20,10))
gs = gridspec.GridSpec(1,2,width_ratios=[10,1])

# First plot
ax1 = plt.subplot(gs[0])
ax1.set_title("Small Piezo Test (sampled every 10ms)")
ax1.set_xlabel("Seconds")
ax1 = sns.lineplot(x=np.linspace(0,10,1000), y = df2['Potential (V)'])

# Second plot
ax2 = plt.subplot(gs[1])
ax2.set_title("Violin plot of sigificant values (>|0.05V|)")
ax2.set_ylabel("Potential (V)")
ax2 = sns.violinplot(y=df2_filtered['Potential (V)'])

plt.show()
```
